## Basic Image Processing 

```python
img = cv2.imread('Shape.jpg')
plt.imshow(img)
```

The code cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) converts the input image (img) from the BGR color space to grayscale using OpenCV. Grayscale images have a single channel, representing the intensity of each pixel, and lack color information. This conversion is commonly used in computer vision and image processing tasks where color information is not necessary, and only the intensity of the pixels is relevant. The resulting grayscale image is stored in the variable grey.

The subsequent plt.imshow(grey) is an attempt to display the grayscale image using matplotlib.pyplot. 

```python
grey =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grey)
```

The code cv2.GaussianBlur(grey, (5, 5), 0) applies a Gaussian blur to the grayscale image (grey) using OpenCV. Here's a brief description:

* cv2.GaussianBlur: This function is used to perform Gaussian blurring on an image.
* grey: The grayscale image to which the blur is applied.
* (5, 5): The size of the Gaussian kernel, which determines the extent of the blurring. In this case, it's a 5x5 kernel.
* 0: The standard deviation of the Gaussian kernel. A value of 0 automatically calculates the standard deviation based on the kernel size.

The purpose of Gaussian blur is to reduce noise in the image and smooth out details. It does this by averaging the pixel values in the vicinity of each pixel, with the weights determined by the Gaussian distribution. The larger the kernel size, the more pronounced the blur.

The resulting blurred image is stored in the variable blurred, and plt.imshow(blurred) is an attempt to display the blurred image using matplotlib.pyplot.

```python
blurred = cv2.GaussianBlur(grey,(5,5),0)
plt.imshow(blurred)
```

The code cv2.Canny(blurred, 50, 150) applies the Canny edge detection algorithm to the blurred image. Here's a concise description:

* cv2.Canny: This function is used for edge detection using the Canny algorithm.
* blurred: The input image on which edge detection is performed.
* 50: The lower threshold for edge gradients. Pixels with gradient values below this threshold are considered non-edges.
* 150: The upper threshold for edge gradients. Pixels with gradient values above this threshold are considered strong edges, and those between the lower and upper thresholds are considered weak edges.

The Canny edge detection algorithm identifies rapid changes in intensity (edges) in an image. It involves several steps, including gradient calculation, non-maximum suppression, and edge tracking by hysteresis. The result is a binary image where edges are highlighted.

The resulting edges are stored in the variable edges, and plt.imshow(edges) is an attempt to display the edges using matplotlib.pyplot. Since the edges image is binary (black and white), no specific colormap is necessary for visualization:

```python
edges = cv2.Canny(blurred,50,150)
plt.imshow(edges)
```

```python
_, thresh = cv2.threshold(grey,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh)
```

Coutours are outlines drawn surrounding the shapes. Its mainly used for shape analysis. cv2.RETR_EXTERNAL means consider only the external aspect of the object, not considering the inside. cv2.CHAN_APPROX_SIMPLE compress horizontal, vertical and diagonal lines and provide a simple averaged representation

```python
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

-1 - How you want the contours to be drawn
(0, 255,0) - colors of the contours

```python
cv2.drawContours(img,contours,-1,(0,255,0),2)
```

```python
def find_triangles(image_path):
    #Read the image
    img =cv2.imread(image_path)
    
    #Convert the image to grayscale 
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    #Apply GaussianBlur to reduce noice and help comtour detection
    blurred = cv2.GaussianBlur(gray,(7,7),0)
    
    #Use canny edge detection to find edges in the image
    edges = cv2.Canny(blurred,30,150)
    
    #Find contours in the edge image 
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    for contour in contours:
        
        #Approximate the contour to a polygon
        epsilon = 0.04 * cv2.arcLength(contour,True)
        approx =cv2.approxPolyDP(contour,epsilon,True)
        
        #If the polygon has 3 vertices its a triangle 
        if len(approx) ==3:
            x, y, w, h = cv2.boundingRect(contour)
            
            #Draw a bounding box around the traingle (in red)
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
            
            
    #Display result
    cv2.imshow('Triangles with Bounding Boxes',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
find_triangles('Shape.jpg')
```
