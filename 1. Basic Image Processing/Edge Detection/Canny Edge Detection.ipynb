```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('E:/CodeBase/GitHub/Image_Processing_and_Computer_Vision/1. Basic Image Processing/Edge Detection/berndsface.png')
plt.imshow(img)
```

```python
blurred = cv2.GaussianBlur(img, (3, 3), 0)
plt.imshow(blurred)
```

```python
edges = cv2.Canny(blurred, 50, 140)
plt.imshow(edges)   
```

```python
_, thresh = cv2.threshold(img, 125, 255, cv2.THRESH_BINARY)
plt.imshow(thresh) 
```
