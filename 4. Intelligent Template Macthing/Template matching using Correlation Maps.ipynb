```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def template_matching(image_path, template_path):

    # Load images (ensure they are grayscale)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE) 
    template = cv2.imread(template_path, cv2.IMREAD_GRAYSCALE)

    # Convert to float32 data type (required for `matchTemplate`)
    image = image.astype(np.float32)
    template = template.astype(np.float32)

    # Subtract means
    image = image - np.mean(image)
    template = template - np.mean(template)

    # Flip the template
    flipped_template = cv2.flip(template, -1)  # Flip both horizontally and vertically

    # Perform template matching using OpenCV's matchTemplate
    correlation_map = cv2.matchTemplate(image, flipped_template, cv2.TM_CCOEFF_NORMED)

    # Display the correlation map
    plt.imshow(correlation_map, cmap='gray')
    plt.colorbar(orientation='horizontal')
    plt.title('Correlation Map')
    plt.axis('off')
    plt.show()

image_path = "E:/CodeBase/GitHub/Image_Processing_and_Computer_Vision/4. Intelligent Template Macthing/church.png"  
template_path = "E:/CodeBase/GitHub/Image_Processing_and_Computer_Vision/4. Intelligent Template Macthing/window.png" 

template_matching(image_path, template_path)
```
